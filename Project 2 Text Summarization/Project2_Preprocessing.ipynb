```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
import string

def preprocess_text(text):
  text = text.lower()
  return text

train_df['text'] = train_df['clean_article'].apply(preprocess_text)
train_df['summary'] = train_df['clean_summary'].apply(preprocess_text)

test_df['text'] = test_df['clean_article'].apply(preprocess_text)
test_df['summary'] = test_df['clean_summary'].apply(preprocess_text)

dev_df['text'] = dev_df['clean_article'].apply(preprocess_text)
dev_df['summary'] = dev_df['clean_summary'].apply(preprocess_text)
```

```python
from nlp_id.lemmatizer import Lemmatizer

lemmatizer = Lemmatizer()

def lemmatize_text(text):
  tokens = word_tokenize(text)
  lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
  return " ".join(lemmatized_tokens)

train_df['text'] = train_df['text'].apply(lemmatize_text)
train_df['summary'] = train_df['summary'].apply(lemmatize_text)

test_df['text'] = test_df['text'].apply(lemmatize_text)
test_df['summary'] = test_df['summary'].apply(lemmatize_text)

dev_df['text'] = dev_df['text'].apply(lemmatize_text)
dev_df['summary'] = dev_df['summary'].apply(lemmatize_text)
```

# Normalization on extractive summary

```python
import ast

# Function to safely convert string representation of list to an actual list of integers
def convert_to_integer_list(column_data):
    if isinstance(column_data, str):
        try:
            # Convert string to list
            result = ast.literal_eval(column_data)
            # Ensure all items are integers
            if all(isinstance(i, int) for i in result):
                return result
            else:
                return []  # Return empty if not all integers
        except (ValueError, SyntaxError):
            return []  # Return empty list if conversion fails
    elif isinstance(column_data, list) and all(isinstance(i, int) for i in column_data):
        return column_data  # Return as is if already a list of integers
    return []  # Return empty if conditions not met

# Apply conversion to the extractive_summary column
train_df['extractive_summary'] = train_df['extractive_summary'].apply(convert_to_integer_list)

# Now we can define the function to rebuild the extractive summary
def rebuild_extractive_summary(clean_article, extractive_summary):
    # Ensure clean_article is a list of lists
    if not isinstance(clean_article, list) or not all(isinstance(sent, list) for sent in clean_article):
        raise ValueError("clean_article should be a list of lists")

    # Ensure extractive_summary contains valid indices
    if not isinstance(extractive_summary, list) or not all(isinstance(idx, int) for idx in extractive_summary):
        raise ValueError("extractive_summary should be a list of integers")

    # Extract sentences based on the indices in extractive_summary
    extracted_sentences = [clean_article[idx] for idx in extractive_summary if idx < len(clean_article)]
    return extracted_sentences

# Apply the function to the entire dataframe
train_df['rebuilt_extractive_summary'] = train_df.apply(
    lambda row: rebuild_extractive_summary(row['clean_article'], row['extractive_summary']),
    axis=1
)

# Display the dataframe to verify the result
print(train_df[['id', 'rebuilt_extractive_summary']].head())
```
